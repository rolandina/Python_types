## Data scrupping with BeautifulSoup

```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = 'https://www.worldometers.info/coronavirus/'
page = requests.get(url)
```

```python
page
```

```python
soup = BeautifulSoup(page.text, 'lxml')
print(soup)
```

```python
table_data = soup.find('table', class_ = 'table table-bordered table-hover main_table_countries')
print(table_data)
# table table-bordered table-hover main_table_countries
# table table-bordered table-hover main_table_countries dataTable no-footer
```

```python
headers = []
for i in table_data.find_all('th'):
    title = i.text
    headers.append(title)

df = pd.DataFrame(columns = headers)

for j in table_data.find_all('tr')[1:]:
        row_data = j.find_all('td')
        row = [tr.text for tr in row_data]
        length = len(df)
        df.loc[length] = row
```

```python
df.head()
```

```python
my_url = "https://en.wikipedia.org/wiki/Harry_Potter_(character)"
```

```python
# opening connection
uClient = uReq(my_url)
```

```python
page_html = uClient.read()
uClient.close()
```

```python
#html parser
page_soup = soup(page_html, "html.parser")
```

```python
labels = page_soup.findAll("th", {"class":"infobox-label"})
data = page_soup.findAll("td", {"class":"infobox-data"})
```

```python
print(labels)
```

```python
headers = []
for i in labels:
    title = i.text
    headers.append(title)
    
```

```python
df = pd.DataFrame(columns = headers)


row = [j.text for j in data]
length = len(df)
df.loc[length] = row
```
